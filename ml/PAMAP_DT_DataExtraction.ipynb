```python
from glob import glob

dat_ls = glob('./data/PAMAP2_Dataset/Protocol/*.dat')
```

```python
dataframe = pd.DataFrame()
for dat_path in dat_ls:
    dataframe = pd.concat([dataframe,
                           pd.read_csv(dat_path, delimiter=' ', header=None)], ignore_index=True)
```

```python
dataframe.head(5)
```

```python
dataframe.iloc[:, 1].value_counts()
```

```python
# dropping activities different than: other, sitting, standing, running, rope jumping
dataframe.drop(dataframe[(dataframe.iloc[:, 1] != 0)
                        & (dataframe.iloc[:, 1] != 2)
                        & (dataframe.iloc[:, 1] != 3)
                        & (dataframe.iloc[:, 1] != 5)
                        & (dataframe.iloc[:, 1] != 24)].index, inplace=True)
```

```python
dataframe.iloc[:, 1].value_counts()
```

```python
# dropping chest and ankle IMU data
dataframe.drop(dataframe.columns[(dataframe.columns > 19)], axis=1, inplace=True)
```

```python
dataframe.columns
```

```python
dataframe.head(5)
```

```python
win_length = 300   # length of time window in ms
undersampling = 10 # add every n-th sample
win_step = win_length//4

progress = IntProgress(min=0, max=len(dataframe)//win_step)
display(progress)

windowed_arr = []
for win_end in range(win_length, len(dataframe), win_step):
    win_start = win_end - win_length
    activity = int(dataframe.iloc[win_start, 1])
    
    features = [activity]
    is_consistent = True
    for j in range(win_start, win_end, undersampling):
        if dataframe.iloc[j, 1] != activity:
            is_consistent = False
            break
            
        if j + undersampling < win_end and abs(dataframe.iloc[j, 0] - dataframe.iloc[j + undersampling, 0]) > undersampling * 0.01 + 0.001:
            is_consistent = False
            break
            
        features = features + list(dataframe.iloc[j, 4:16])
    
    if is_consistent:
        windowed_arr.append(features)
        
    progress.value += 1
```

```python
windowed_df = pd.DataFrame(windowed_arr)
```
